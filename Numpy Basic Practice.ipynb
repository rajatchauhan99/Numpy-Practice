2. Let's look at the speed of numpy:

```python
A = np.arange(1000)
```

```python
B = np.arange(1000)
```

```python
start_time = time.time()
result1 = A+B                 
end_time = time.time()
print("Execution time is: ", (end_time - start_time)*1000 )
```

```python
result1
```

```python
A1 = range(1000)
B1 = range(1000)
```

```python
start_time = time.time()
result2 = [(x+y) for x,y in zip(A1,B1)]
end_time = time.time()
print("Execution time is: ", (end_time - start_time)*1000 )
```

so we can see execution time of numpy array to do the same stuff is around 25 percent
